```python
import pandas as pd
url = 'data_crab.csv'
dataframe = pd.read_csv(url)
data = dataframe.drop(['sp', 'sex', 'index'], axis = 'columns')
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
data.plot(kind = 'scatter', x = 'CL', y = 'CW',figsize = (9,9), color = 'black')
```

```python
#Mo hinh bac 2
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame([data['CL'],data['CL']**2]).T, y = data['CW'])
print(model.intercept_)
print(model.coef_)
model.score(X = pd.DataFrame([data['CL'],data['CL']**2]).T, y = data['CW'])*100 
```

```python
import numpy as np
range = np.arange(10,50,1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'CL', y = 'CW', figsize = (9,9), color = 'black')
plt.plot(range, y_pred, color = 'red')
```
