```python
import pandas as pd
url = 'Housing.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist() #So lan lap lai cua data
plt.show()
```

```python
#Su tuong quan giua cac dai luong
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind= 'scatter', x = 'area', y = 'price', figsize = (6 , 6), color = 'blue')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['area']), y= data['price'])
print(model.intercept_)
print(model.coef_)
```

```python
#Do chinh xac
model.score(X = pd.DataFrame(data['area']), y= data['price'])
```

```python
data.plot(kind= 'scatter', x = 'area', y= 'price', figsize = (5,5), color = 'red')
y_pred = model.predict(X = pd.DataFrame(data['area']))
plt.plot(data['area'], y_pred, color = 'blue')
```
